```python
import pandas as pd
import csv

df = pd.read_csv('likely_handwritten_files.tsv',sep='\t', quoting=csv.QUOTE_NONNUMERIC).drop('Unnamed: 0',
                                                                                                    axis=1)
df.head()
```

```python
# Load HTRed texts from handwritten files


import glob, shutil

PATH_PREFIX = '/ivi/ilps/personal/vprovat/Valkema_HTR_results/'

for filename in glob.glob('/ivi/ilps/personal/vprovat/transkribus_results/transkribus_results/*/txt/*'):
    name = filename.split('_')[-1]
    shutil.move(filename, PATH_PREFIX+name)
```

```python
df['path_to_text'] = df['images.reference'].apply(lambda name:
                                                 PATH_PREFIX + name.split('\\')[-1].split('.')[0] + '.txt'
                                                           )
```

```python
# Places
offset = df_letters['places'].isnull().idxmax() # first nan index, 0 at first
step = 100

while step + offset < len(df_letters):
    if isinstance(df_letters['places'][offset],set): # already processed
        continue 
    df_letters['places'][offset:step+offset] = df_letters['path_to_text'][offset:step+offset].progress_map(get_places_by_path)
    df_letters.to_csv('handwritten_letters_with_entities_partial.tsv',sep='\t')
    offset = step + offset
    
df_letters['places'][offset:len(df_letters)] = df_letters['path_to_text'][offset:len(df_letters)].progress_map(get_places_by_path)
df_letters.to_csv('handwritten_letters_with_entities.tsv',sep='\t')
```
