```python
# People
# offset = 2100 # todo: set manually
offset = df_typed_letters['people'].isnull().idxmax() # first nan index
step = 100 # 13-15 mins per batch with 100, is this faster?


while step + offset < len(df_typed_letters):
    print(offset)
#     if isinstance(df_typed_letters['people'][offset],set): # already processed
#         offset = step + offset
#         continue 
    df_typed_letters['people'][offset:step+offset] = df_typed_letters['path_to_text'][offset:step+offset].progress_map(get_people_by_path)
    df_typed_letters.to_csv('typed_letters_with_entities_final.tsv',sep='\t',index=False)
    offset = step + offset
    
df_typed_letters['people'][offset:len(df_typed_letters)] = df_typed_letters['path_to_text'][offset:len(df_typed_letters)].progress_map(get_people_by_path)
df_typed_letters.to_csv('typed_letters_with_entities_final.tsv',sep='\t',index=False)
    
```

```python
# Places
offset = 0
step = 100

while step + offset < len(df_typed_letters):
    if df_typed_letters['places'][offset:step+offset]: # already processed
        continue 
    df_typed_letters['places'][offset:step+offset] = df_typed_letters['path_to_text'][offset:step+offset].progress_map(get_places_by_path)
    df_typed_letters.to_csv('typed_letters_with_entities_partial.tsv',sep='\t')
    offset = step + offset
    
df_typed_letters['places'][offset:len(df_typed_letters)] = df_typed_letters['path_to_text'][offset:len(df_typed_letters)].progress_map(get_places_by_path)
df_typed_letters.to_csv('typed_letters_with_entities.tsv',sep='\t')
```
